```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(programming_languages, popularity, color=['red', 'green', 'blue', 'yellow', 'purple',
'orange'])
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
import matplotlib.pyplot as plt

programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

plt.pie(popularity, labels=programming_languages)

plt.title('Popularity of Programming Languages')

plt.show()
```

```python
import matplotlib.pyplot as plt
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]
plt.scatter(math_marks, science_marks)
plt.xlabel('Mathematics Marks')
plt.ylabel('Science Marks')
plt.title('Comparison of Mathematics and Science Marks')
plt.show()
```
